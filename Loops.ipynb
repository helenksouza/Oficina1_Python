## Range

são utilizados para gerar sequências numéricas de forma não aleatória, mas sim de maneira específica. 



```python
for num in range(1,10): 
  print(num)
```

```python
for num in range(10): #padrão: inicia em 0 e termina em 9
  print(num)
```

```python
for num in range(1,10, 2): #padrão: inicia em 0 e termina em 9
  print(num)
```

```python
for num in range(10,1, -1): #padrão: inicia em 0 e termina em 9
  print(num)
```

```python
lista = list(range(10))
lista
```

## Loop while

while: expressão booleana

```python
num= int(input('Insira um numero'))

while num < 5: #loo infinito
  print(num)

```

```python
resposta=input('Insira sua resposta!')

while resposta != 'sim':
  resp=input('Você vai?')
```

```python
while True:
  comando = input('digite sair para sair: ')
  if comando == 'sair':
    break
```
